```python
import numpy as np
from sklearn.metrics import roc_auc_score
from pprint import pprint

# Assume 4 classes (e.g., ICBHI: 0=normal, 1=crackle, 2=wheeze, 3=both)
n_classes = 4
y_true = np.random.randint(0, n_classes, size=200)
y_pred = np.random.randint(0, n_classes, size=200)

# simulate model probabilities (softmax outputs)
y_prob = np.random.rand(200, n_classes)
y_prob = y_prob / y_prob.sum(axis=1, keepdims=True)

metrics = compute_classification_metrics(y_true, y_pred, y_prob, n_classes)
pprint(metrics)
```
